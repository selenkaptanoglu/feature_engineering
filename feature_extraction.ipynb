```python
import pandas as pd
```

```python
df=pd.read_csv('/content/titanic.csv')
df.head()
```

```python
#Letter Count
df['Name_Count']=df["Name"].str.len()
df.head()
```

```python
#Word Count
df["Name_Word_Count"]=df["Name"].apply(lambda x:len(str(x).split(" ")))
df.head()
```

```python
# ozel yapıları yakalamak
df["Name_Mr"]=df["Name"].apply(lambda x: len([x for x in x.split() if x.startswith("Mr")]))
df.head()
```

```python
df["Name_Dr"]=df["Name"].apply(lambda x: len([x for x in x.split() if x.startswith("Dr")]))

df.groupby("Name_Dr").agg({"Survived":["mean","count"]})
```

```python
from datetime import date

dff=pd.read_csv('/content/course_reviews.csv')
dff.head()
```

```python
dff.info()
```

timestamp obje türünde önce date olarak veri tipini değiştirelim

```python
dff['Timestamp'] = pd.to_datetime(dff["Timestamp"], format="%Y-%m-%d %H:%M:%S")
```

```python
# year
dff['year'] = dff['Timestamp'].dt.year
```

```python
# month
dff['month']=dff['Timestamp'].dt.month
```

```python
#year diff
dff['year_diff']=date.today().year - dff['Timestamp'].dt.year
```

```python
#month diff(iki tarih arasındaki ay farkı): yıl farkı + ay farkı
dff['month-diff']=(date.today().year - dff['Timestamp'].dt.year)*12+date.today().month - dff['Timestamp'].dt.month
```

```python
#day name
dff['day_name']=dff['Timestamp'].dt.day_name()
```
